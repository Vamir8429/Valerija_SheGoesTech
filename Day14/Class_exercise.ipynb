```python
# create a code cell for all imports that you need
import random
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# create a code cell for all variables that you need
dice = 4
throw = 100
```

```python
# have a cell that runs your previous homework from day 14
# ideally you would see a graph in the notebook
def dice_throw2(dice, throw):
    dice_list = [random.randint(1, 4) for _ in range(0,throw)]
    print(dice_list)
    #plot a histogram of the list
    plt.hist(dice_list,bins=range(dice+1,throw+1))
    plt.xlabel('Throw')
    plt.ylabel('Dice')
    plt.title('Histogram of Dice Throws')
    plt.show()

dice_throw2(3,10)
```
