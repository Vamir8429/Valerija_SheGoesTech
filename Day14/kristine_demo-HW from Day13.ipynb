```python
import matplotlib.pyplot as plt
import datetime as dt
import random
from collections import Counter
import pandas as pd # Excel on steroids
import matplotlib.pyplot as plt
import seaborn as sns # prettier plots for matplotlib
```

```python
def single_throw(number_of_dice=4):
    """Throw a number of dice and return the sum of the dice"""
    # i could add some safety checks here such as number_of_dice > 0
    return sum([random.randint(1, 6) for _ in range(number_of_dice)])
```

```python
single_throw()
```

```python
def dice_throws(dice_num, throws):
    """Throw a number of dice a number of times and return the results"""
    return [single_throw(dice_num) for _ in range(throws)]
```

```python
dice_throws(2, 10)
```

```python
def plot_dices(dices_list, file_name='', show_bar_values=False):
    count_list = Counter(dices_list)
    df = pd.DataFrame.from_records(list(dict(count_list).items()), columns=['dices','count'])
    # print(df) # printing should be optional for larger workloads
    chart = sns.barplot(data=df, x='dices', y='count', color='purple')
    if show_bar_values:
        for i in chart.containers:  #to display bar values
            chart.bar_label(i,)
    
    chart.set(xlabel='Dices')
    chart.set(ylabel='Count')
    chart.set(title=f'Total Number of Occurence of Each Dice')
    date = dt.datetime.now()
    plt.suptitle(date, ha='left', va='bottom', size='x-small')
    if file_name:
        plt.savefig(file_name)

    return dices_list, plt.show(), file_name
```

```python
# if __name__ == "__main__": # ok but not needed in a cell

# so a good looking pipeline would be:
# we get data
my_dice_list = dice_throws(6, 100_000)
# we plot the data
my_plot = plot_dices(my_dice_list)
```

```python
plot_dices(dice_throws(6, 100_000), 'Dices_Histogram.jpg')
```
